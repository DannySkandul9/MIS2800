```python
import sqlite3
```

```python
conn = sqlite3.connect('NFL_Scouting.db')
```

```python
cursor = conn.cursor()
```

```python
print("Database connected successfully")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS scouting_stats (
    team_name TEXT,
    city TEXT,
    wins INTEGER,
    points_per_game REAL
)
""")
```

```python
conn.commit()
```

```python
print("Table created successfully")
```

```python
cursor.execute("""
INSERT INTO scouting_stats (team_name, city, wins, points_per_game)
VALUES
    ('Chiefs', 'Kansas City', 11, 29.2),
    ('Bills', 'Buffalo', 10, 26.5),
    ('Cowboys', 'Dallas', 12, 27.8),
    ('Jets', 'New York', 7, 18.6)
""")
```

```python
conn.commit()
```

```python
print("Data inserted successfully")
```

```python
cursor.execute("""
SELECT * FROM scouting_stats
WHERE points_per_game > 24
""")
results = cursor.fetchall()
for row in results:
    print(row)
```

```python
results = cursor.fetchall()
```

```python
print("Teams averaging more than 24 points per game:")
for row in results:
    print(row)
```

```python
cursor.execute("""
SELECT * FROM scouting_stats
ORDER BY wins DESC
""")
results_sorted = cursor.fetchall()
for row in results_sorted:
    print(row)
```

```python
results = cursor.fetchall()
```

```python
print("Teams sorted by wins (highest to lowest):")
for row in results:
    print(row)
```

```python
cursor.execute("""
SELECT * FROM scouting_stats
WHERE wins > 5
ORDER BY city ASC
""")
results_filtered_sorted = cursor.fetchall()
for row in results_filtered_sorted:
    print(row)
```

```python
results = cursor.fetchall()
```

```python
print("Teams with more than 5 wins, sorted by city:")
for row in results:
    print(row)
```

```python
conn.close()
```
